```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import math
import sklearn
import sklearn.datasets

from opt_utils import load_params_and_grads, initialize_parameters, forward_propagation, backward_propagation
from opt_utils import compute_cost, predict, predict_dec, plot_decision_boundary, load_dataset,relu,sigmoid
from testCases import *

%matplotlib inline
plt.rcParams['figure.figsize'] = (7.0, 4.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
def load_params_and_grads(seed=1):
    
    # 初始化参数
    np.random.seed(seed)
    W1 = np.random.randn(2,3)
    b1 = np.random.randn(2,1)
    W2 = np.random.randn(3,3)
    b2 = np.random.randn(3,1)
    
    # 初始化梯度参数
    dW1 = np.random.randn(2,3)
    db1 = np.random.randn(2,1)
    dW2 = np.random.randn(3,3)
    db2 = np.random.randn(3,1)

    return W1, b1, W2, b2, dW1, db1, dW2, db2

def initialize_parameters(layer_dims):
    """
    输入：
        layer_dims：是各个层的神经元的个数组成的数组
    """
    np.random.seed(3)
    parameters = {}
    # 获得数组的程度
    L = len(layer_dims)
    
    # 从第二层开始
    for l in range(1,L):
        # 为了获得前一层的维度，需要减1
        parameters['W'+str(l)] = np.random.randn(
            layer_dims[l],layer_dims[l-1])*0.01
        parameters['b'+str(l)] = np.zeros((
            layer_dims[l],1))
        
    return parameters

def forward_propagation(X,parameters):
    
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    W3 = parameters['W3']
    b3 = parameters['b3']
    
    #进行前向运算
    #线性->relu->线性->relu->线性->sigmoid
    z1 = np.dot(W1,X) + b1
    a1 = relu(z1)
    z2 = np.dot(W2,a1) + b2
    a2 = relu(z2)
    z3 = np.dot(W3,a2) + b3
    a3 = sigmoid(z3)
    
    cache = (z1, a1, W1, b1, z2, a2, W2, b2, z3, a3, W3, b3)
    
    return a3, cache

def backward_propagation(X,Y,cache):

    m = X.shape[1]
    (z1, a1, W1, b1, z2, a2, W2, b2, z3, a3, W3, b3) = cache
    
    dz3 = 1./m*(a3-Y)
    dw3 = np.dot(dz3,a2.T)
    db3 = np.sum(dz3,axis=1,keepdims=True)
    
    da2 = np.dot(W3.T,dz3)
    dz2 = np.multiply(da2,np.int64(a2>0))    # 对relu求导计算
    dW2 = np.dot(dz2,a1.T)
    db2 = np.sum(dz2,axis=1,keepdims=True)
    
    da1 = np.dot(W2.T,dz2)
    dz1 = np.multiply(da1,np.int64(a1>0))
    dW1 = np.dot(dz1,X.T)
    db1 = np.sum(dz1,axis=1,keepdims=True)
    
    gradients = {"dz3": dz3, "dW3": dW3, "db3": db3,
             "da2": da2, "dz2": dz2, "dW2": dW2, "db2": db2,
             "da1": da1, "dz1": dz1, "dW1": dW1, "db1": db1}

    return gradients
    
def compute_cost(a3,Y):
    
    m = Y.shape[1]
    
    logprobs = np.multiply(-np.log(a3),Y) + np.multiply(-p.log(1-a3),1-Y)
    cost = 1./m * np.sum(logprobs)
    
    return cost

def predict(X,y,parameters):
    
    m = X.shape[1]
    p = np.zeros((1,m),dtype=np.int)
    
    a3,cache = forward_propagation(X,parameters)
    
    for i in range(0,a3.shape[1]):
        if a3[0,i] >0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0
            
            
    print("Accuracy: "  + str(np.mean((p[0,:] == y[0,:]))))
    
    return p

def predict_dec(parameters, X):
    
    # Predict using forward propagation and a classification threshold of 0.5
    a3, cache = forward_propagation(X, parameters)
    predictions = (a3 > 0.5)
    return predictions


def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=np.squeeze(y), cmap=plt.cm.Spectral)
    plt.show()

def load_dataset():
    np.random.seed(3)
    train_X, train_Y = sklearn.datasets.make_moons(n_samples=300, noise=.2) #300 #0.2 
    # Visualize the data
    plt.scatter(train_X[:, 0], train_X[:, 1], c=train_Y, s=40, cmap=plt.cm.Spectral);
    train_X = train_X.T
    train_Y = train_Y.reshape((1, train_Y.shape[0]))

    return train_X, train_Y
```

### Batch梯度下降与随机梯度下降

Batch Gradient Descent 方法是指将所有的m个样本输入完训练后，完成一次参数更新。
Stochastic Gradient Descent(随机梯度下降)：是指每一个输入数据训练后，就执行一次参数更新。

关于参数更新的公式如下：
![](imgs/2.jpg)

- l=1,...,L表示对应层的number

对应的coding如下：

```python
def update_parameters_with_gd(parameters, grads, learning_rate):
    """
    输入：
        parameters：缓存的参数字典
            parameters['W' + str(l)] = Wl
            parameters['b' + str(l)] = bl
        grads: 缓存的梯度字典
            grads['dW' + str(l)] = dWl
            grads['db' + str(l)] = dbl
        learning_rate：学习率
    """
    L = len(parameters)//2
    
    for l in range(L): # 从零开始，而参数是从1开始W1
        parameters["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate * grads["dW" + str(l+1)]
        parameters["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate * grads["db" + str(l+1)]
        
    return parameters
```

### Mini-Batch 梯度下降

我们来看看如何对数据进行批量梯度下降，主要有两步：

- Shuffle:对样本数据进行洗牌重新排列，这样就可以保证数据的随机性。这里需要注意的是，记得要将label数据也跟着一起重新排列。
过程如下图所示：
![](imgs/5.jpg)

- Partition:对数据进行批量化，每一批数据可以聚集mini_batch_size个数据。
![](imgs/6.jpg)

这里有一个需要注意的问题，就是当数据批量化后，最后一组数据，如果m个样本无法被mini_batch_size整除，那么剩下来的数据计算公式如下：
$m - {mini\_batch\_size} x |\frac m{mini\_batch\_size}|$

所以我们可以实践如下coding：

**一个numpy知识点**
函数shuffle与permutation都是对原来的数组进行重新洗牌
- shuffle直接在原来的数组上进行操作，改变原来数组的顺序，无返回值
- permutation不直接在原来的数组上进行操作，而是返回一个新的打乱顺序的数组，并不改变原来的数组


```python
# 关于shuffle与permutation实践
m = 100
arr = np.arange(1,100,1)
print(arr)
np.random.shuffle(arr)
print(arr)
print(np.random.permutation(m))
```

```python

def random_mini_batches(X, Y, mini_batch_size = 64, seed = 0):
    np.random.seed(seed)
    m = X.shape[1]
    mini_batches = []
    
    # 第一步，数据洗牌重排
    permutation = list(np.random.permutation(m))
    # 更新洗牌的数据
    shuffled_X = X[:,permutation]
    shuffled_Y = Y[:,permutation].reshape((1,m))
    
    #第二步，数据批量组
    # 计算可以分成多少批数据
    num_complete_minibatches = math.floor(m/mini_batch_size)
    for k in range(0,num_complete_minibatches):
        mini_batch_X = shuffled_X[:,k*mini_batch_size:(k+1)*mini_batch_size]
        mini_batch_Y = shuffled_Y[:,k*mini_batch_size:(k+1)*mini_batch_size]
        
        mini_batch = (mini_batch_X,mini_batch_Y)
        mini_batches.append(mini_batch)
    # 最后的一批数据的集合
    if m % mini_batch_size != 0:
        mini_batch_X = shuffled_X[:,num_complete_minibatches*mini_batch_size:m]
        mini_batch_Y = shuffled_Y[:,num_complete_minibatches*mini_batch_size:m]
        
        mini_batch = (mini_batch_X,mini_batch_Y)
        mini_batches.append(mini_batch)
        
    return mini_batches
```

### Momentum优化算法

我们知道使用了mini-batch 批量样本输入后，数据的梯度的更新会变得更加平滑，但是还是具有一定的摆动，这种摆动导致收敛的速度会降低，所以我们这里使用Momentum优化算法，进一步减少这样的摆动，加快数据收敛。

从数学上来了解释Momentum方法：
- 使用了指数加权平均的思想，让更多的历史数据加入到梯度更新中来
- 通过更多的历史数据这样可以抵消一些摆动
- 这里加入$v$这个变量
- 你可以想象成加速的v，帮助数据加速收敛

为了获得这个v的数据，首先需要对dW,db进行赋值为零的初始化。


```python
def initialize_velocity(parameters):
    L = len(parameters) // 2 # number of layers in the neural networks
    v = {}

    # 初始化
    for l in range(L):
        
        v["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        v["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)


    return v
```

Momentum数学公式如下：
![](imgs/7.jpg)
式中：
- l: 表示第l层网络的数据1，...,L
- $\beta$:表示momentum动量的值
    - 其值为(0,1)
    - 如果看过指数加权平均，就可以知道，其代表了使用了多少个前面的数据进行加权
- $\alpha$表示学习率

下面我们来实现这个coding:

```python
def update_parameters_with_momentum(parameters, grads, v, beta, learning_rate):
    
    L = len(parameters)//2
    
    for l in range(L):
        # 计算v的值
        v["dW" + str(l+1)] = beta*v["dW"+str(l+1)] + (1-beta)*grads["dW"+str(l+1)]
        v["db" + str(l+1)] = beta*v["db"+str(l+1)] + (1-beta)*grads["db"+str(l+1)]
        # 更新参数
        parameters["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate*v["dW" + str(l+1)]
        parameters["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate*v["db" + str(l+1)]
    
    return parameters,v
```

从上面的coding我们可以知道，由于开始的时候初始化了v为0，所以开始的时候，数据更新摆动会比较大，但是到了后面，v的值迭代一定次数后，基本可以进行稳定的平均，将会获得一个比较好的结果。



### Adam

这个优化算法集合了Momentum与RMSProp算法的优点，所以该算法在很多场合都获得较好的效果。
具体公式如下：
![](imgs/8.jpg)

式中：
- $v$:表示Momentum的参数
- $v^{corrected}$表示修正后的参数
- $s$:表示RMSProp的参数
- $s^{corrected}$表示修正后的参数
- t :表示执行了第几个批量数据
- L:层数
- $\beta_1$ 与 $\beta_2$两个指数加权平均超参数
- $\alpha$表示学习率
- $\epsilon$ 为了避免被除数为0

下面对v,s 进行初始化

```python
def initialize_adam(parameters):
    L = len(parameters) // 2 # number of layers in the neural networks
    v = {}
    s = {}

    # 初始化
    for l in range(L):
        
        v["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        v["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)
        s["dW" + str(l+1)] = np.zeros(parameters["W" + str(l+1)].shape)
        s["db" + str(l+1)] = np.zeros(parameters["b" + str(l+1)].shape)

    return v,s
```

```python

def update_parameters_with_adam(parameters, grads, v, s, t, learning_rate = 0.01,
                                beta1 = 0.9, beta2 = 0.999,  epsilon = 1e-8):
    L = len(parameters)//2
    v_corrected = {}    
    s_corrected = {}  
    
    for l in range(L):
        # 计算v的值
        v["dW" + str(l+1)] = beta1 * v["dW" + str(l+1)] + (1 - beta1) * grads['dW' + str(l+1)]
        v["db" + str(l+1)] = beta1 * v["db" + str(l+1)] + (1 - beta1) * grads['db' + str(l+1)]
        
        # 计算修正后的v值
        v_corrected["dW" + str(l+1)] = v["dW" + str(l+1)] / (1 - beta1 ** t)
        v_corrected["db" + str(l+1)] = v["db" + str(l+1)] / (1 - beta1 ** t)
        
        
        # 计算s值
        s["dW" + str(l+1)] = s["dW" + str(l+1)] + (1 - beta2) * (grads['dW' + str(l+1)] ** 2)
        s["db" + str(l+1)] = s["db" + str(l+1)] + (1 - beta2) * (grads['db' + str(l+1)] ** 2)


        #计算修正后的s值
        s_corrected["dW" + str(l+1)] = s["dW" + str(l+1)] / (1 - beta2 ** t)
        s_corrected["db" + str(l+1)] = s["db" + str(l+1)] / (1 - beta2 ** t)
       
        # 更新参数
        parameters["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate * ( v_corrected["dW" + str(l+1)] / (np.sqrt(s_corrected["dW" + str(l+1)]) + epsilon))
        parameters["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate * ( v_corrected["db" + str(l+1)] / (np.sqrt(s_corrected["db" + str(l+1)]) + epsilon))

    return parameters,v,s
```

下面我们开始定义该模型：

```python
def model(X, Y, layers_dims, optimizer, learning_rate = 0.0007, mini_batch_size = 64, beta = 0.9,
          beta1 = 0.9, beta2 = 0.999,  epsilon = 1e-8, num_epochs = 10000, print_cost = True):
    
    
    # 获得层数
    L = len(layers_dims)
    #记录cost的值
    costs = []
    #记录第几批数据
    t = 0
    # 保证随机值
    seed =10
    
    # 初始化参数
    parameters = initialize_parameters(layers_dims)
    
    # 根据不同的优化方法，初始参数
    if optimizer == "gd":
        pass
    elif optimizer == "momentum":
        v = initialize_velocity(parameters)
    elif optimizer == "adam":
        v,s = initialize_adam(parameters)
    
    for i in range(num_epochs):
        seed = seed + 1
        
        # 获得批量数据
        minibatches = random_mini_batches(X,Y,
                                          mini_batch_size,
                                          seed)
        for minibatch in minibatches:
            # 选取一个minibatch的数据样本
            (minibatch_x,minibatch_y) = minibatch
            
            #前向传播
            a3,caches = forward_propagation(minibatch_x,parameters)
            # loss值
            cost = compute_cost(a3,minibatch_y)
            
            # 反向传播计算梯度
            grads = backward_propagation(minibatch_x,
                                        minibatch_y,caches)
            
            # 更新参数
            if optimizer == "gd":
                parameters = update_parameters_with_gd(parameters,
                                                      grads,
                                                      learning_rate)
            elif optimizer == "momentum":
                parameters,v = update_parameters_with_momentum(parameters,
                                                               grads,
                                                                v,
                                                              beta,
                                                              learning_rate)
            elif optimizer == "adam":
                parameters,v,s = update_parameters_with_adam(parameters,
                                                            grads,
                                                            v,
                                                            s,
                                                            t,
                                                            learning_rate,
                                                            beta1,
                                                            beta2,
                                                            epsilon)
       # 打印cost
        if print_cost and i % 1000 == 0:
            print ("Cost after epoch %i: %f" %(i, cost))
        if print_cost and i % 100 == 0:
            costs.append(cost)

    # plot the cost
    plt.plot(costs)
    plt.ylabel('cost')
    plt.xlabel('epochs (per 100)')
    plt.title("Learning rate = " + str(learning_rate))
    plt.show()
    
    return parameters
            
```
