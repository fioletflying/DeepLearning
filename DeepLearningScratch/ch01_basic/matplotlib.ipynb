## Matplotlib 简介

在深度学习的实验中，图形的绘制和数据的可视化非常重要。Matplotlib 是
用于绘制图形的库，使用 Matplotlib 可以轻松地绘制图形和实现数据的可视化。

### 绘制简单的图形

- np.arange(0,6,0.1) ： 自动生成对应数据列
- plt.plot(x,y)： 绘制图像的函数


```python
import numpy as np
import matplotlib.pyplot as plt

#  以0.1为单位，生成0到6的数据
x = np.arange(0,6,0.1) 
# sin值计算
y = np.sin(x)

#绘制x,y对应的图像
plt.plot(x,y)
# 显示该图像
plt.show()
```

### pyplot具体使用

pyplot 的添加标题和 x 轴标签名等其他功能

```python
import numpy as np
import matplotlib.pyplot as plt

#  以0.1为单位，生成0到6的数据
x = np.arange(0,6,0.1) 
# sin值计算
y1 = np.sin(x)
# cos值计算
y2 = np.cos(x)

#绘制x,y对应的图像
# lable 该线的标签名，用于在图中说明该线代表的含义，linestyle:线的形状
plt.plot(x,y1,label="sin",linestyle="--")
plt.plot(x,y2,label="cos",linestyle="dotted")
# x,y的标签
plt.xlabel("x")
plt.ylabel("y")
# 图形的标题
plt.title("sin & cos")
# 显示标签
plt.legend()
# 显示该图像
plt.show()
```

### 图像的显示

pyplot 中还提供了用于显示图像的方法 imshow()。另外，可以使用
matplotlib.image 模块的 imread() 方法读入图像。

关于更多的matplotlib的操作可以查看[matplotlib](https://www.matplotlib.org.cn/)



```python
# %matplotlib inline”就是模仿命令行来访问magic函数的在IPython中独有的形式
# 由于 %matplotlib inline 的存在，当输入plt.plot()后，不必再输入 plt.show()，图像将自动显示出来
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread("bits.jpg")
plt.imshow(img)
plt.show()
```
