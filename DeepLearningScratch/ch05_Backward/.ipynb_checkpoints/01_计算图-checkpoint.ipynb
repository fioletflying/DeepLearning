## 计算图

计算图就将数学计算的过程用图形来表示，里面有节点和箭头线来表示，将整个的计算过程和数据的流动形象的表示出来。

### 用计算图求解问题

这里举一个实际的日子。
我们在超市买了 2 个 100 元一个的苹果，消费税是10%，请计算
支付金额。

我们就可以绘制如下的计算图：
![](imgs/1.jpg)

上面图中，用节点来表示运算符，数据用线条来表示。

所以计算图的流程就是：
- 构建计算图
- 从左往右进行计算

这就是正向传播的过程（forward propagation）
那么如果计算图从右往左计算，这个过成就是反向传播（backward propagation）

### 计算图的优点

#### 局部计算

计算图可以集中精力于局部计算。无论全局的计算有多么复杂，各个步骤所要做的就是对象节点的局部计算。虽然局部计算非常简单，但是通过传递它的计算结果，可以获得全局的复杂计算的结果。

#### 将中间的计算结果全部保存起来

从上面的计算过程我们可以看到，我们计算的每一个步骤的计算结果都会保存起来，有利于我们检查。


#### 反向传播高效计算导数

这个优点是我们使用计算的最大原因，其实就是我们所说的链式求导。

### 链式法则

链式法则是用来计算复合函数的求导问题.这里有一个实际的例子:
$z=(x+y)^2$构成的式子分解如下:  
$
z=t^2
$

$
t=x+y
$

**如果某个函数由复合函数表示，则该复合函数的导数可以用构成复合函数的各个函数的导数的乘积表示。**
