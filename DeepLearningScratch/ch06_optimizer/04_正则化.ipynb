## 正则化

过拟合现象的直观理解，模型可以很好的拟合训练数据的状态，但是不能很好的拟合训练数据以外的数据状态。
所以为了提高模型的泛化能力，也就是可以很好预测没有包含在训练集里的数据。
**所以我们需要使用正则化来提高数据的泛化能力，解决过拟合的问题。**

### 过拟合的原因

主要有两个原因：
- 训练数据少
- 模型参数较多，表现力强

下面我们来看个实验，这里
- 使用MNIST 300个数据来训练 （训练数据少）
- 使用7层网络（模型参数较多）

最后我们可以看到的结果如下：
![](imgs/23.jpg)

从图中我们可以看到现象是：
- 训练精度高，但是测试精度较差。
说明出现了过拟合的现象。

### 权重衰减（L2 正则化）

权重衰减就是在训练的过程中对大的权重进行惩罚，来防止过拟合。这里的一个假设是：

- 很多过拟合原本就是因为权重参数取值过大才发生的。

这里我们的做法就是在损失函数里面增加一项，权重的平方范数（L2范数），这样就可以抑制权重变大，
![](imgs/24.jpg)
- L2 范数 $\frac 12 \lambda W^2$
- $ \lambda$ 就是一个超参数来控制正则化的强度
- 求导L2范数则变成$\lambda W$
- 在反向传播的过程中会加上这个梯度值，从而减少W的值，抑制权重参数过大

通过上面的变换后，我们得到的数据测试图表如下：

![](imgs/25.jpg)

通过正则化后：
- 两者的精度差距变小了，这说明过拟合受到了抑制

### Dropout

Dropout 是一种在学习的过程中随机删除神经元的方法,。训练时，随机选出隐藏层的神经元，然后将其删除。被删除的神经元不再进行信号的传递.具体的过程如下图：

![](imgs/26.jpg)

#### 前向传播中的dropout

这里我们需要处理的是一个三层的模型，一般在输出层最后一层我们不使用dropout，具体的步骤如下：

- 创建一个随机值d,该随机值的shape与a的一样
- 将d中的值与keep_prob进行比较，获得一个标志位，这样就可以知道那些节点失效
- 将a= a\*d,获得最终的有效的节点和无效的节点
- a= a/keep_prob,保证损失值得期望值

#### dropout的反向传播

反向传播的过程中，只要将前向传播中的失效的节点，继续让其失效即可。具体步骤如下：

- 将前向传播中的失效的节点，继续让其失效 D * A
- 由于前向传播中，A1/keep_prob,所以在反向传播中也需要将da/keep_prob
