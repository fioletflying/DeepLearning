## 简单感知机的原理

### 感知机的概念

#### 为什么要理解感知机
感知机就是人工神经元或者叫朴素感知机。理解了这里的简单感知机可以让我们很好的理解以后比较复杂的深度学习原理的底层。因为两者的底层很多原理是相通的。

#### 什么是感知机

感知机就是一个信号处理的单元(节点，中转站)
- 输入可以是多个信号源
- 输出是一个信号

这里我们更加简单的理解输出的信号：这里我们经过处理后，信号输出只有两种状态：
- 输出：1
- 输出：0

#### 感知机的结构

如下图所示
![感知机的结构示意图](imgs/1.jpg)

感知机的组成：
- 输入信号源：x1,x2
- 输出信号：y
- 每个信号的权重：w1,w2

感知机如何处理这些信息：
- 信号汇总：y = x1*w1+x2*w2
- 信号判断：y 与阈值$\theta$比较
    - 当y> $\theta$,神经源被激活，输出为1
    - 当y< $\theta$,神经源不激活，输出为0
    
具体的函数表示如下:
![](imgs/2.jpg)

感知机的多个输入信号都有各自固有的权重，这些权重发挥着控制各个信号的重要性的作用。也就是说，权重越大，对应该权重的信号的重要性就越高。

### 简单的逻辑电路

如何利用感知机来解决我们实际中遇到的问题呢？这里我们先从简单的逻辑电路入手。

#### 与门

与门是有两个输入和一个输出的门电路。当两个信号输入都为真时输出就为1，其他的为0。 如下图所示：
![与门](imgs/3.jpg)

如何利用感知机来实现这个与门的逻辑电路。这里我们知道输入以及对应的输出值。所以我们需要将权重w1，w2，以及阈值的设定
这里我们可以很多的选择来实现这个电路：  
比如：（w1=0.5,w=0.5,$\theta=0.6$）（w1=0.3,w=0.3,$\theta=0.5$）等等

#### 与非门和或门

与非门的意思是，与上面的与门真好相反，两个为真时则0，其他则为1.如下图所示：  
![与非门](imgs/4.jpg)

感知机实现，可以设置的参数为(w1=-0.5,w=-0.5,$\theta=-0.6$)

或门的意思是，两个输入有一个为真则输出为1，其他则为0.如下图所示：
![或门](imgs/5.jpg)

感知机实现，可以设置的参数为(w1=0.5,w=0.5,$\theta=0.4$)


这里我可以看到这些参数是通过输入和输出的结果，我们人工设计出来的。以后，我们需要做的就是，这些参数是机器自动
学习出来，不需要人工干预。这就机器学习的过程。
