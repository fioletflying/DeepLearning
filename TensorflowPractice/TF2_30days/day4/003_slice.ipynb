```python
import tensorflow as tf
import numpy as np

print(tf.__version__)
```

```python
# 创建一个4维张量
# 可以将其认为是4张图片信息，size 32x32,通道为RGB 3 通道
x = tf.random.normal([4,32,32,3])
```

```python
# 获取第一张图片
print(x[0])
```

```python
# 取第 1 张图片的第 2 行
print(x[0][1])
```

```python
# 取第 1 张图片，第 2 行，第 3 列的数据
x[0][1][2]
```

```python
# 取第 3 张图片，第 2 行，第 1 列的像素， B 通道(第 2 个通道)颜色强度值
x[2][1][0][1]
```

```python
# 也可以直接用逗号隔开来获取
#  取第 2 张图片，第 10 行，第 3 列的数据
x[1,9,2]
```

```python
# 读取第 2,3 张图片
x[1:3]
```

```python
# 读取第一张图片
x[0,::]
```

```python
# 获取所有图片中前28行和28列中偶数行，列的数据
x[:,0:28:2,0:28:2,:]
```
