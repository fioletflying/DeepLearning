```python
import tensorflow as tf
import numpy as np

print(tf.__version__)
```

```python
# 创建一个4维张量
# 可以将其认为是4张图片信息，size 32x32,通道为RGB 3 通道
x = tf.random.normal([4,32,32,3])
```

```python
# 获取第一张图片
print(x[0])
```

```python
# 取第 1 张图片的第 2 行
print(x[0][1])
```

```python
# 取第 1 张图片，第 2 行，第 3 列的数据
x[0][1][2]
```

```python
# 取第 3 张图片，第 2 行，第 1 列的像素， B 通道(第 2 个通道)颜色强度值
x[2][1][0][1]
```

```python
# 也可以直接用逗号隔开来获取
#  取第 2 张图片，第 10 行，第 3 列的数据
x[1,9,2]
```

```python
# 读取第 2,3 张图片
x[1:3]
```

```python
# 读取第一张图片
x[0,::]
```

```python
# 获取所有图片中前28行和28列中偶数行，列的数据
# 连续冒号中的第二间隔的数量 1表示：没有间隔，2：间隔为2，
x[:,0:28:2,0:28:2,:]
```

```python
x = tf.range(9)
print(x)
# 逆序取值，间隔为-1 表示从后往前取值
print(x[8:0:-1])
# 逆序取值的简写
print(x[::-1])
# 逆序间隔取值
print(x[::-2])
```

```python
x = tf.random.normal([4,32,32,3])
# 行、列逆序间隔采样
x[0,::-2,::-2] 
```

```python
# 取某个通道的值
x[:,:,:,1]
# 读取第 1~2 张图片的 G/B 通道数据
# 高宽维度全部采集
x[0:2,...,1:]
```

```python
# 读取 R/G 通道数据
# 所有样本，所有高、宽的前 2 个通道
# 用省略号表示前面的数据通道所有样本，这样避免写那么多冒号
# x[:,:,:,:2]
x[...,:2]
```

### tf.gather

从params的axis维根据indices的参数值获取切片,适合抽取不连续区域的子集.



```python
# tf.gather
# data: [classes, students, subjects]
a = tf.random.normal([4,35,8])
# 获取第2，3个班级的数据
print(tf.gather(a,axis=0,indices=[2,3]).shape)
# 该索引下的数据
print(tf.gather(a,axis=0,indices=[2,1,3,0]).shape)

# 获取4个班级中，第2,3,7,9,11学生的成绩
print(tf.gather(a,axis=1,indices=[2,3,7,9,11]).shape)

# 获取4个班级中，所有学生中第3，5，7,工3门课的成绩
print(tf.gather(a,axis=2,indices=[3,5,7]).shape)
```
